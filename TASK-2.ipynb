## Exploration of numerical variables

### Distribution of Numerical Features


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
train_data_path = r'C:\Users\Satyaki Nandi\Desktop\jupyter\train.csv'
train_df = pd.read_csv(train_data_path)

# Handling missing values
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
train_df['Fare'].fillna(train_df['Fare'].median(), inplace=True)

# List of numerical features
numerical_features = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived', 'Pclass']

# Plot histograms
train_df[numerical_features].hist(bins=15, figsize=(15, 10), layout=(3, 2), edgecolor='black')
plt.suptitle('Distribution of Numerical Features')
plt.show()

```

### Correlation Matrix

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = train_df[numerical_features].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

### Pair Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Pair plot
sns.pairplot(train_df, vars=numerical_features, hue='Survived', diag_kind='kde')
plt.suptitle('Pair Plots of Numerical Features by Survival', y=1.02)
plt.show()
```

### Box Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Box plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.boxplot(x='Survived', y='Age', data=train_df, ax=axes[0, 0])
axes[0, 0].set_title('Age vs Survived')

sns.boxplot(x='Survived', y='Fare', data=train_df, ax=axes[0, 1])
axes[0, 1].set_title('Fare vs Survived')

sns.boxplot(x='Survived', y='SibSp', data=train_df, ax=axes[1, 0])
axes[1, 0].set_title('Siblings/Spouses Aboard vs Survived')

sns.boxplot(x='Survived', y='Parch', data=train_df, ax=axes[1, 1])
axes[1, 1].set_title('Parents/Children Aboard vs Survived')

plt.tight_layout()
plt.show()
```

### Violine Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Violin plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.violinplot(x='Survived', y='Age', data=train_df, ax=axes[0, 0])
axes[0, 0].set_title('Age vs Survived')

sns.violinplot(x='Survived', y='Fare', data=train_df, ax=axes[0, 1])
axes[0, 1].set_title('Fare vs Survived')

sns.violinplot(x='Survived', y='SibSp', data=train_df, ax=axes[1, 0])
axes[1, 0].set_title('Siblings/Spouses Aboard vs Survived')

sns.violinplot(x='Survived', y='Parch', data=train_df, ax=axes[1, 1])
axes[1, 1].set_title('Parents/Children Aboard vs Survived')

plt.tight_layout()
plt.show()
```

## Survival analysis of the dataset

### Survival Rate by Gender

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Survival rate by gender
plt.figure(figsize=(5,5))
sns.barplot(x='Sex', y='Survived', data=train_df)
plt.title('Survival Rate by Gender')
plt.show()

# Display actual counts
print(train_df.groupby('Sex')['Survived'].value_counts(normalize=True).unstack())
```

### Survival Rate by Passenger Class

```python
# Survival rate by passenger class
plt.figure(figsize=(5,5))
sns.barplot(x='Pclass', y='Survived', data=train_df)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Display actual counts
print(train_df.groupby('Pclass')['Survived'].value_counts(normalize=True).unstack())
```

###  Survival Rate by Age

```python
# Survival rate by age
plt.figure(figsize=(12,6))
sns.histplot(train_df, x='Age', hue='Survived', element='step', stat='density', common_norm=False, kde=True)
plt.title('Survival Rate by Age')
plt.show()
```

### Survival Rate by Embarkation Port

```python
# Survival rate by port of embarkation
plt.figure(figsize=(5,5))
sns.barplot(x='Embarked', y='Survived', data=train_df)
plt.title('Survival Rate by Port of Embarkation')
plt.show()

# Display actual counts
print(train_df.groupby('Embarked')['Survived'].value_counts(normalize=True).unstack())
```

### Combined Factor Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Survival rate by gender and passenger class
plt.figure(figsize=(10, 6))
sns.catplot(x='Pclass', y='Survived', hue='Sex', data=train_df, kind='bar')
plt.title('Survival Rate by Gender and Passenger Class')
plt.show()
```
